```python
import pandas as pd 
import os 
```

```python
df = pd.read_csv("./avg_int_per_cell_data.csv")

agg = (
    df.groupby(["well_id", "FOV"], as_index=False)
      .agg({
          "nuclei_avg_int": "mean",
          "cytoplasm_avg_int": "mean",
      })
)

agg["nuc_to_cyt_ratio"] = agg["nuclei_avg_int"] / agg["cytoplasm_avg_int"]

agg

# Define the .csv path
csv_path = "avg_int_per_well_id.csv"

# Append to the .csv with new data points each round
agg.to_csv(csv_path, mode="a", index=False, header=not os.path.isfile(csv_path))
```
